```python
def removal_html_tags(text):
    pattern = "<.*?>"
    clean_text = re.sub(pattern,"",text)
    return clean_text
```

```python
def remove_space(text):
    pattern = "\s+"
    clean_text = re.sub(pattern," ",text)
    return clean_text
```

```python
df["new_text"] = df["new_text"] .apply(remove_space)
```

```python
def count_punctuation(text):
    count = 0
    for char in exclude:
        if char in text:
            count = count+1
    return count
```

```python
fake_char_count = df[df["label"]==0]["char_count"]
true_char_count = df[df["label"]==1]["char_count"]
```

```python
#estimate avg char count for spam and ham--->CLT
def estimate_CLT(data,sample_size,no_of_samples):
    mean_lst=[]
    for i in range(no_of_samples):
        smaples=np.random.choice(data,size=sample_size)
        mean_lst.append(np.mean(smaples))
    mean_Array=np.array(mean_lst)
    return np.mean(mean_Array)
```

```python
sample_size=30
no_of_samples=1000
data=fake_char_count
avg_spam_char=estimate_CLT(data,sample_size,no_of_samples)
```

```python
avg_spam_char
```

```python
sample_size=30
no_of_samples=1000
data=true_char_count
avg_spam_char=estimate_CLT(data,sample_size,no_of_samples)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()

params_grid_knn = {
    "n_neighbors": [3, 5, 7, 9, 11]
}

grid_knn = GridSearchCV(
    estimator=knn_model,
    param_grid=params_grid_knn,
    cv=5,
    scoring="accuracy"
)

grid_knn.fit(xtrain_tfidf, ytrain)

print("Best parameters for KNN:", grid_knn.best_params_)
print("Best score for KNN:", grid_knn.best_score_)

best_knn_model = KNeighborsClassifier(n_neighbors=grid_knn.best_params_['n_neighbors'])
best_knn_model.fit(xtrain_tfidf, ytrain)
ypred_knn = best_knn_model.predict(xtest_tfidf)

print("KNN Accuracy Score:", accuracy_score(ytest, ypred_knn))
print("\nKNN Classification Report:\n", classification_report(ytest, ypred_knn))
print("\nKNN Confusion Matrix:\n", confusion_matrix(ytest, ypred_knn))

```

```python
from sklearn.svm import SVC

svm_model = SVC()

params_grid_svm = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"]
}

grid_svm = GridSearchCV(
    estimator=svm_model,
    param_grid=params_grid_svm,
    cv=3,
    scoring="accuracy",
    verbose=2,
    n_jobs=-1
)

grid_svm.fit(xtrain_tfidf, ytrain)

print("Best parameters for SVM:", grid_svm.best_params_)
print("Best score for SVM:", grid_svm.best_score_)

best_svm_model = SVC(C=grid_svm.best_params_['C'], kernel=grid_svm.best_params_['kernel'])
best_svm_model.fit(xtrain_tfidf, ytrain)
ypred_svm = best_svm_model.predict(xtest_tfidf)

print("\nSVM Accuracy Score:", accuracy_score(ytest, ypred_svm))
print("\nSVM Classification Report:\n", classification_report(ytest, ypred_svm))
print("\nSVM Confusion Matrix:\n", confusion_matrix(ytest, ypred_svm))
```
